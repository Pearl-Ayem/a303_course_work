## Gravity Correction for Eco-Celli






```python
LAT=49.2606
MEAN_HEIGHT=5.182   #meters ASL
gn=9.80665          #standard acceleration of gravity m/s^2 
Hp= 105             # station elevation in gpm; 
Td = 273.15 + 4.1  # dew point temperature in K averaged between 11am to 1pm
T= 4.53
TS= 273.15 + T     #station temperature in K; 
lr= 0.0065         #the assumed lapse-rate in the fictitious air column extending
                    #from sea level to the level of the station elevation level  K/gpm; 
a= 0.61
r= 0.6113*math.exp((2.5e6/461)*((1/273.15)-(1/Td)))
RH= 0.9711
T_virtual = T*(1+(a*r)) 
T_time_before = 3.4

    
def spatial_gravity(lat,height,mean_height):
    h = (height-mean_height)
    h = (1.118e-6)*h
    h = ((3.086e-6)*height)+h
    g = 1-((2.6373e-3)*math.cos(2*lat))+((5.9e-6)*(math.cos(2*lat)**2))
    g = 9.80616*g
    g = g-h
    return g
    
g=spatial_gravity(LAT,Hp,MEAN_HEIGHT)
Tv_star=0.5*(T_virtual+T_time_before+(Hp*0.0065))
Tv_star=273.15+Tv_star 
slr=math.exp(Hp/(29.3*Tv_star))

def sea_level_reduction(p):
    return p*slr

def gravity_cor(lat,height,mean_height,raw):
    P= (g*raw)/gn
    return P
        
def inHG_to_kPa(x):
    return (x*101.325)/29.92126

def hPa_to_kPa(x):
    return (x*101.325)/1013.25
```

```python
df['rmy']=df['rmy'].apply(sea_level_reduction)
df['ec'] = df['ec'].apply(inHG_to_kPa)
df['yvr'] = df['yvr'].apply(inHG_to_kPa)
df['k4'] = df['k4'].apply(hPa_to_kPa)
df['k5'] = df['k5'].apply(hPa_to_kPa)
df['esb'] = df['esb'].apply(hPa_to_kPa)
df_raw=df.copy()
ax = df.plot(x='time')
ax.set_title('Raw Data (all measurements converted to kPa)')
ax.set_xlabel('Time')
ax.set_ylabel('Pressure (kPa)')
ax.legend(['EcoCelli','Kestrel 4000','Kestrel 5000','RM Young','ESB','YVR'])
plt.xticks(rotation=60)
```

```python
df['ec'] = gravity_cor(LAT,Hp,MEAN_HEIGHT,df.ec)
df['ec']=df['ec'].apply(sea_level_reduction)
df_cal=df.copy()
ax = df.plot(x='time')
ax.set_title('Calibrated Data')
ax.set_xlabel('Time')
ax.set_ylabel('Pressure (kPa)')
ax.legend(['EcoCelli','Kestrel 4000','Kestrel 5000','RM Young','ESB','YVR'])
plt.xticks(rotation=60)
```

We assume teh index correction, i = 0 because the Eco-Celli was calibrated to the temperature before the lab began, making it a safe assumption that the index correction factor i is 0. 

As for the temperature correction, I assumed this to be 0 as well because the Eco-Celli was not filled with mercury, while all of the corrections are with mercury as a function of temperature. These temperature effects must be compensated, which they were by the initial calibration between the pressure and temperature. 

The mean height was estimated at 5.182 m, as per the geoid height from the EGM96 model. The 105m height ASL for UBC is assumed as per the calibration of the Kestrels.

The sea level reduction was done using the formula given lecture:



**Mixing Ratio is found through the following formula:**

$mixing\;ratio\;(r) = relative\;humidity * saturated\;mixing\;ratio \LARGE$

This relationship between saturation vapor pressure and temperature is approximated by the Clausius-Clapeyron equation:

$ saturated\;mixing\;ratio(e_s)=e_0*exp[\frac{L}{R}*(\frac{1}{T_0}-\frac{1}{T_d})]$


$ saturated\;mixing\;ratio(e_s)=0.6113\;kPa*exp[\frac{2.5e6\;Jkg^-1}{461\;Jkg^-1K^-1}*(\frac{1}{273.15\;K}-\frac{1}{T_d})] $
